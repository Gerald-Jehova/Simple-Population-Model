```python
import pandas as pd
import altair as alt
```

```python
age = [1, 2, 3]
n =  [25, 5, 5]
sx = [0.8, 0.5, 0]
bx = [1, 3, 2]
#bx = [2, 2.7, 0.1]
s0 = 0.4
t = 20

global all_pops
all_pops = []
def compute(age,n,sx,bx,s0, t):
  global all_pops
  names = ['Age', 'n' + str(1), 'Surviv', 'offspr', 'offsurv', 'New n']
  for q in range(t):
    new_n = []
    newborns = []
    newborns_survive = []
    preshift = []
    for i in range(len(age)):
      newborns.append(n[i] * bx[i])
      newborns_survive.append(newborns[i]*s0)
      new_n.append(n[i] * sx[i])
    preshift = new_n.copy()
    new_n.insert(0, sum(newborns_survive))
    new_n.pop()
    print(f'{names[0]:^7}|{names[1]:^8}|{names[2]:^8}|{names[3]:^8}|{names[4]:^7}|{names[5]:^8}')
    for i in range(3):
      print(f'{age[i]:6.2f} | {n[i]:6.2f} | {preshift[i]:6.2f} | {newborns[i]:6.2f} | {newborns_survive[i]:5.2f} | {new_n[i]:5.2f}')
    print(f'  Sum:{sum(n):9.2f}{sum(newborns_survive):26.2f}{sum(new_n):8.2f}')
    print()
    print()
    names[1] = 'n' + str(q + 2)
    all_pops.append(n)
    n = new_n

info = ['Age', 'n(i)', 's(x)', 'b(x)', 's(0)']
print(f'{info[0]:^7}|{info[1]:^8}|{info[2]:^8}|{info[3]:^8}|{info[4]:^8}')
for i in range(len(age)):
    if(i == 0):
        print(f'{age[i]:6.2f} | {n[i]:6.2f} | {sx[i]:6.2f} | {bx[i]:6.2f} | {s0:5.2f}')
    else:
        print(f'{age[i]:6.2f} | {n[i]:6.2f} | {sx[i]:6.2f} | {bx[i]:6.2f} |')
print()
print()
compute(age, n, sx, bx, s0, t)
```

```python
pops = pd.DataFrame()
```

```python
new = []
for i in all_pops:
    new.extend(i)
pops['individuals'] = new
```

```python
years = []
counter = 0
for i in range(len(new)):
    if(i % 3  == 0):
        counter += 1
    years.append(counter)
pops['Year'] = years
```

```python
ages = []
for i in range(len(new)):
    if(i % 3 == 0):
        counter = 1
    else:
        counter += 1
    ages.append(counter)
pops['Age'] = ages
```

```python
counter = 0
totals = []
for i in range(len(all_pops)):
    for q in range(len(all_pops[i])):
        totals.append(sum(all_pops[i]))
pops['Total'] = totals
```

```python
pops['Proportion'] = pops['individuals'] / pops['Total']
pops['Lambda'] = ''
for i in pops.index:
    if(i-3 < 0):
        continue
    pops.loc[i,'Lambda'] = pops.loc[i,'Total'] / pops.loc[i-3,'Total']
pops
```

```python
chart = alt.Chart(pops).mark_bar().encode( 
	x = 'Year:O', 
	y = alt.Y("sum(individuals):Q", title="Total Population"),
	color = 'Age:N'
).properties(
    title = 'Population by Age and Year'
)
chart
```
